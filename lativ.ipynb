```python
#同種類的每一件衣服
import os
import json
import urllib
import shutil #檔案下載與儲存的函式庫
import requests
from bs4 import BeautifulSoup

#建立資料夾
dirname="lativ_skirt" #資料夾名字(要改)
os.mkdir(dirname)

dd = 0 #服飾編號
x = 1  #檔名流水號 (要改)
cloth_id = [] #存href (/Detail/52091011)
img_url = []  #存圖片網址 (https://s1.lativ.com.tw/i/52091/5209101_A_01.jpg)
j = {}  #json



res = requests.get('https://www.lativ.com.tw/WOMEN/bottoms/skirt')  #發送請求給網頁 
soup = BeautifulSoup(res.text,'html.parser') #解析爬回來的頁面資料 指定解析模式

results = soup.find_all("a", {"class": "imgd"})

for result in results:
    if 'href' in result.attrs:
        cloth_id.append(result['href'])

        

for _id in range(len(cloth_id)):
    if cloth_id[_id][8:14] != dd:   #挑選出同款不同色的
        dd = cloth_id[_id][8:14]
        url = "https://www.lativ.com.tw" +cloth_id[_id]
        res1 = requests.get(url)  #對子分頁送出請求
        soup1 = BeautifulSoup(res1.text,'html.parser')  #解析爬回來的頁面資料 指定解析模式   
       
        j["brand"] = "lativ"
        j["porduct_name"] = soup1.h1.text  #json porduct_name
        j["gender"] = "w"  #(要改)
        
        price = soup1.find("span", {"id": "price"})
        j["price"] = price.text
        
        j["website_url"] = url
        div = soup1.find('div', {"class": "oldPic show"})
        _a = div.find_all("a")
   
        for a in _a:
           # print(a)
            for img in a:
                if 'data-original' in img.attrs:
                    img_url = img['data-original']
                 
                    
                    image_save_path = './' +dirname+ '/lative_%d.jpg' % x     #檔名:lative_流水號.jpg
                    urllib.request.urlretrieve(img_url,image_save_path)       #下載圖片
                
                    with open('./' +dirname+ '/lative_%d.json' %x, 'w') as fp:  #將所得數據存為json文件
                        json.dump(j, fp = fp, ensure_ascii = False, indent = 1, sort_keys = True)
                 
                    x += 1
                
                if 'src' in img.attrs:
                    if img['src'].endswith('.jpg'):
                        img_url = img['src']  
                    
                        
                        image_save_path = './' +dirname+ '/lative_%d.jpg' % x
                        urllib.request.urlretrieve(img_url,image_save_path)
                        
                        with open('./' +dirname+ '/lative_%d.json' %x, 'w') as fp: 
                            json.dump(j, fp = fp, ensure_ascii = False, indent = 1, sort_keys = True)
                        
                        x += 1
```
