```python
import os
import json
import urllib
import shutil #檔案下載與儲存的函式庫
import requests
from bs4 import BeautifulSoup
import numpy


#女 下身
path = 'D:/styler/img/UR/'
j = {}  #json
repeat = []
h = []
m='1' #圖片、json流水號 記得要改

for page in range(1,2):#1頁
    url = 'https://www.urban-research.tw/category/item/page/x/?item=all&brand=all&type=women&itemcat=04bottom&sort=new&num=60&key'
    url = url.replace('x',str(page))#前者替換為後者 page是每次迴圈提取資料 str是強制轉換為文字型態(以這個例子只跑1、2 兩次)
    res = requests.get(url)
    soup = BeautifulSoup(res.text,'html.parser')#
    results = soup.find_all("a")
    for r in results:
        if 'href' in r.attrs:
            url = r['href']
            if url.endswith('#1') or url.endswith('#2') or url.endswith('#3') or url.endswith('#4') or url.endswith('#5')  :
                href1 = url.split('/')[3].split('#')[0]
                if href1 not in h:
                    h.append(href1)
#                     print(href1)
                    res1 = requests.get(url)  #對子分頁送出請求
                    soup1 = BeautifulSoup(res1.text,'html.parser')
                    
                    #品牌
                    j['brand'] = 'URBAN RESEARCH'
                    
                
                    #品名
                    for name in soup1.find_all('h1',{'class':'item_name'}):
                        name = name.text
#                         print(name)
                        j['porduct_name'] = name
                     
                    #性別
                    j['gender'] = 'W'
                    
                    #價錢
                    for price in soup1.find_all('p',{'class':'item_info_price'}):
                        price = price.text
                        price = price.split('元')[0]
                        price = price.replace(',','')
#                         print(price)
                        j['price'] = price
                      
                    
                    #網址
                    j['website_url'] = url
                    
                    
                    #存圖片
                    for save in soup1.find_all('a',{'class':'thumb'}):
                        if save not in repeat:
                            repeat.append(save)
                            print(save['href'])
                            m = m.zfill(5)
                            image_save_path = path + 'URBAN RESEARCH_'+ str(m) +'.jpg'  
                            print('save path',image_save_path)
                            print(j)
#                             urllib.request.urlretrieve(save['href'],image_save_path)
#                             with open(path+ 'URBAN RESEARCH_'+str(mm)+'.json', 'w') as fp:         #將所得數據存為json文件
#                                 json.dump(j, fp = fp, ensure_ascii = False, indent = 4, sort_keys = True)
                            m=int(m)
                            m+=1
                            m=str(m)
                           
                    
                    
    
```
