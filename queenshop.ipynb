```python
import os
import json
import urllib
import shutil #檔案下載與儲存的函式庫
import requests
from bs4 import BeautifulSoup

#建立資料夾
path = 'D:/styler/img/queenshop/top/'
j = {}  #json
repeat = []
href = []
m='1' #圖片、jaon流水號 記得要改

website = 'https://www.queenshop.com.tw/zh-TW/QueenShop/'
for page in range(1,2):#1頁
    url = 'https://www.queenshop.com.tw/zh-TW/QueenShop/ProductList?item1=04&item2=qs&item3=01&Page=x&View=4'
    url = url.replace('x',str(page))#前者替換為後者 page是每次迴圈提取資料 str是強制轉換為文字型態(以這個例子只跑1、2 兩次)
    res = requests.get(url)
    soup = BeautifulSoup(res.text,'html.parser')#
    results = soup.find_all("li")
    j["brand"] = "QUEENSHOP"

 
    for result in results:
        r = result.find('a')
        if 'href' in r.attrs:
            if 'Product?item' in  r['href']:
                rr = website+r['href']
#                 print('r',rr)
                href.append(rr)  #rr就是website_url
                res1 = requests.get(rr)  #對子分頁送出請求
                soup1 = BeautifulSoup(res1.text,'html.parser')
                             
                #紀錄品名
                ttt = soup1.find_all('h4')
                for ttt1 in ttt:
                        if '有些超值單品不要錯過' not in ttt1:
#                             print('name',ttt1.text)
                            pn = ttt1.text
                            j['product_name'] = pn
                            if '十色售' in pn:
                                pn = pn.replace('十色售','')
                                j['product_name'] = pn
                            if '八色售' in pn:
                                pn = pn.replace('八色售','')
                                j['product_name'] = pn
                            if '六色售' in pn:
                                pn = pn.replace('六色售','')
                                j['product_name'] = pn
                            if '五色售' in pn:
                                pn = pn.replace('五色售','')
                                j['product_name'] = pn
                            if '四色售' in pn:
                                pn = pn.replace('四色售','')
                                j['product_name'] = pn
                            if '三色售' in pn:
                                pn = pn.replace('三色售','')
                                j['product_name'] = pn
                            if '兩色售' in pn:
                                pn = pn.replace('兩色售','')
                                j['product_name'] = pn
                            if 'S/M/L' in pn:
                                pn = pn.replace('S/M/L','')
                                j['product_name'] = pn
                            if ' ' in pn:
                                pn = pn.replace(' ','')
                                j['product_name'] = pn
                        
                        
                        
                
                #性別
                j["gender"] = "w" 
                  
                #價錢
                tt = soup1.find_all('span',{'name':'p_original_d'})
                for tt1 in tt:
                        p = tt1.text
                        p = p.replace('NT. ','')
#                       print('price',p)
                        j["price"] = p 
                    
                    
                #網址
                j["website_url"] = str(rr)
                
                #存圖片
                t = soup1.find_all('img' )
                for t1 in t:
                    if 'data-src' in t1.attrs:
                        u = t1['data-src']
                        if u not in repeat:
                            print('src',t1['data-src'])
                            repeat.append(u)
                            m = m.zfill(5)
                            image_save_path = path + 'QUEENSHOP_'+ str(m) +'.jpg'  
                            print('save path',image_save_path)
                            print(j)
#                             urllib.request.urlretrieve(t1['data-src'],image_save_path) #存檔
#                             with open(path+ 'QUEENSHOP_'+str(mm)+'.json', 'w') as fp:         #將所得數據存為json文件
#                                 json.dump(j, fp = fp, ensure_ascii = False, indent = 4, sort_keys = True)
                            m=int(m)
                            m+=1
                            m=str(m)  
                           
               
```
